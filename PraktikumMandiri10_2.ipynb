```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
# Data dari tabel
data = {
    "NIM": ["TI001","TI002","TI003","TI004","TI005",
            "TI006","TI007","TI008","TI009","TI010"],
    "Hasil Sebenarnya": ["Lulus","Lulus","Lulus","Lulus","Lulus",
                         "Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus"],
    "Hasil Prediksi": ["Lulus","Lulus","Lulus","Tidak Lulus","Tidak Lulus",
                       "Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus","Tidak Lulus"]
}

df = pd.DataFrame(data)
df
```

```python
# Ubah menjadi binary (Opsional, tapi memudahkan perhitungan)
df["true"] = df["Hasil Sebenarnya"].map({"Lulus":1, "Tidak Lulus":0})
df["pred"] = df["Hasil Prediksi"].map({"Lulus":1, "Tidak Lulus":0})

df
```

```python
# Confusion matrix
cm = confusion_matrix(df["true"], df["pred"])
print("Confusion Matrix:")
print(cm)
```

```python
# Metrics
accuracy  = accuracy_score(df["true"], df["pred"])
precision = precision_score(df["true"], df["pred"])
recall    = recall_score(df["true"], df["pred"])

print("\nAccuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
```

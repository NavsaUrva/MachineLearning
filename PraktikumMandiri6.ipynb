```python
# Import library pandas untuk membaca data
import pandas as pd

# Membaca dataset Iris dari folder drive
df = pd.read_csv(path + "/data/BreastCancerWisconsin.csv")

# Menampilkan 5 data pertama untuk memastikan dataset berhasil dibaca
df
```

```python
# Daftar kolom yang mau di-drop (kalau 'Unnamed: 32' ngga ada di datasetmu, ga masalah)
drop_cols = ['id', 'diagnosis']
if 'Unnamed: 32' in df.columns:
    drop_cols.append('Unnamed: 32')

# Fitur numerik untuk model
X = df.drop(columns=drop_cols)

# Target klasifikasi
y = df['diagnosis']

# Cek dimensi fitur dan target
print("Dimensi X:", X.shape)
print("Dimensi y:", y.shape)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Contoh mapping: {'B':0, 'M':1}
print("Mapping label -> angka:")
for kelas, kode in zip(le.classes_, le.transform(le.classes_)):
    print(kelas, "->", kode)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

print("Jumlah data train:", len(X_train))
print("Jumlah data test :", len(X_test))
```

```python
from sklearn.svm import SVC

# Membuat model SVM dengan kernel linear
svm_model = SVC(kernel='linear', random_state=42)

# Melatih model menggunakan data training
svm_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Prediksi label untuk data testing
y_pred = svm_model.predict(X_test)

# Akurasi
print(f"Akurasi: {accuracy_score(y_test, y_pred) * 100:.2f}%")

# Laporan klasifikasi (precision, recall, f1-score untuk tiap kelas)
# kita tampilkan nama kelas aslinya (B/M) biar gampang dibaca
print("\nLaporan Klasifikasi:\n")
print(classification_report(
    y_test,
    y_pred,
    target_names=le.classes_   # urutannya sama dengan encoding tadi
))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(4,3))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Reds",
    xticklabels=le.classes_,   # Predicted label names
    yticklabels=le.classes_    # True label names
)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - SVM (Breast Cancer)")
plt.show()
```

```python
# === 3D Scatter: Perbandingan Label Asli vs Hasil Prediksi (Breast Cancer) ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401
from sklearn.preprocessing import LabelEncoder

# 1) Pilih 3 fitur untuk divisualisasikan
f1, f2, f3 = 'radius_mean', 'texture_mean', 'perimeter_mean'

# 2) Buat encoder BARU yang di-fit ke label string (B/M) pada seluruh df
le_full = LabelEncoder()
df['DiagnosisEncoded'] = le_full.fit_transform(df['diagnosis'])  # 'B'/'M' -> 0/1 (urut sesuai le_full.classes_)

# 3) Siapkan warna & nama kelas (mengikuti urutan le_full)
class_names = list(le_full.classes_)      # contoh: ['B', 'M']
palette = ['g', 'r'][:len(class_names)]   # 2 warna cukup

# 4) Pastikan prediksi (y_pred) selaras:
#    - SVC sudah dipakai dengan label numerik (0/1), biasanya sudah oke.
#    - Simpan prediksi hanya untuk indeks test agar tidak rancu.
df['PredictedEncoded'] = np.nan
df.loc[X_test.index, 'PredictedEncoded'] = y_pred  # y_pred diharapkan 0/1

# Jika ternyata y_pred masih string (jarang), map dulu ke 0/1:
if df.loc[X_test.index, 'PredictedEncoded'].dtype == 'object':
    df.loc[X_test.index, 'PredictedEncoded'] = le_full.transform(df.loc[X_test.index, 'PredictedEncoded'])

# 5) Plot 3D: kiri = label asli (seluruh data), kanan = label prediksi (test set)
fig = plt.figure(figsize=(18, 7))

# (A) Label asli
ax1 = fig.add_subplot(121, projection='3d')
for i, cls in enumerate(class_names):
    subset_true = df[df['DiagnosisEncoded'] == i]
    ax1.scatter(
        subset_true[f1], subset_true[f2], subset_true[f3],
        color=palette[i], label=cls, s=40, alpha=0.85
    )
ax1.set_title('Label Asli (Seluruh Data)')
ax1.set_xlabel(f1); ax1.set_ylabel(f2); ax1.set_zlabel(f3)
ax1.legend(title='Diagnosis')

# (B) Label prediksi (khusus test set)
ax2 = fig.add_subplot(122, projection='3d')
test_df = df.loc[X_test.index].dropna(subset=['PredictedEncoded'])
for i, cls in enumerate(class_names):
    subset_pred = test_df[test_df['PredictedEncoded'] == i]
    ax2.scatter(
        subset_pred[f1], subset_pred[f2], subset_pred[f3],
        color=palette[i], label=f'Predicted: {cls}', s=40, alpha=0.85
    )
ax2.set_title('Label Prediksi (Test Set)')
ax2.set_xlabel(f1); ax2.set_ylabel(f2); ax2.set_zlabel(f3)
ax2.legend(title='Diagnosis')

plt.suptitle('Perbandingan Label Asli vs Hasil Prediksi (3D Scatter)', fontsize=14)
plt.show()
```

```python
# === 3D Plot hasil klasifikasi per kelas ===
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 (aktifkan backend 3D)
import matplotlib.pyplot as plt

# Encode label string 'diagnosis' -> angka (0/1) untuk pewarnaan
le3d = LabelEncoder()
df['diagnosis_encoded'] = le3d.fit_transform(df['diagnosis'])
labels = le3d.classes_          # contoh: array(['B', 'M'], dtype=object)
colors = ['tab:blue', 'tab:red'] # warna untuk tiap kelas (B, M)

# Pilih 3 fitur yang representatif untuk visualisasi
feat_x, feat_y, feat_z = 'radius_mean', 'texture_mean', 'perimeter_mean'

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot tiap kelas dengan warna berbeda
for i, nama_kelas in enumerate(labels):
    subset = df[df['diagnosis_encoded'] == i]
    ax.scatter(
        subset[feat_x], subset[feat_y], subset[feat_z],
        color=colors[i], label=str(nama_kelas), s=50, alpha=0.7
    )

ax.set_xlabel(feat_x.replace('_', ' '))
ax.set_ylabel(feat_y.replace('_', ' '))
ax.set_zlabel(feat_z.replace('_', ' '))
ax.set_title("3D Visualisasi Klasifikasi Breast Cancer (SVM)")
ax.legend()
plt.show()
```

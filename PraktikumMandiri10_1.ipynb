```python
import pandas as pd
import math
```

```python
data = {
    "Temperatur Udara (°C)": [10, 25, 15, 20, 18, 20, 22, 24],
    "Kecepatan Angin (km/jam)": [0, 0, 5, 3, 7, 10, 5, 6],
    "Klasifikasi / Persepsi Marry": [
        "Dingin", "Panas", "Dingin", "Panas",
        "Dingin", "Dingin", "Panas", "Panas"
    ]
}
```

```python
df = pd.DataFrame(data)

df
```

```python
# --- Data testing ---
temp_test = 16   # suhu 16°C
angin_test = 3   # kecepatan angin 3 km/jam
```

```python
# --- Hitung selisih dan jarak (Euclidean) ---
df["selisih suhu"] = (df["Temperatur Udara (°C)"] - temp_test) ** 2
df["selisih angin"] = (df["Kecepatan Angin (km/jam)"] - angin_test) ** 2
df["jarak"] = (df["selisih suhu"] + df["selisih angin"]).apply(math.sqrt)

df
```

```python
# Contoh k = 3
k = 5
k_neighbors = df.head(k)
print("\nK =", k)
print(k_neighbors[["Temperatur Udara (°C)", "Kecepatan Angin (km/jam)",
                   "Klasifikasi / Persepsi Marry", "jarak"]])
```

```python
# Contoh k = 5 (untuk kasus tie 2:2 → jadi 3:2)
k = 5
k_neighbors_5 = df.head(k)
prediksi_k5 = k_neighbors_5["Klasifikasi / Persepsi Marry"].mode()[0]
print("Prediksi dengan k=5:", prediksi_k5)
```
